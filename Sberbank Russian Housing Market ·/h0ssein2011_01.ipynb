```python
import pandas as pd
import numpy as np
import zipfile
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
zf_train = zipfile.ZipFile('train.zip') 
zf_macro = zipfile.ZipFile('macro.zip') 
train = pd.read_csv(zf_train.open('train.csv'))
macro=pd.read_csv(zf_macro.open('macro.csv'))
merged=train.merge(macro,on='timestamp')
```

```python
merged.head()
```

```python
na_count=(merged.isnull().sum()/merged.shape[0])*100
na_count.sort_values(ascending=False)
```
